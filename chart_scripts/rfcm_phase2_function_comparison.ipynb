```python
import numpy as np
import matplotlib.pyplot as plt

def plot_function_comparison():
    # 設定距離範圍 (X軸: 距離 d)
    d = np.linspace(0, 4.5, 500)
    
    # 1. 傳統歐氏距離誤差 (Traditional Euclidean Error)
    # J 正比於 d^2 (平方成長)
    y_euclidean = d**2
    
    # 2. RFCM 第二階段指數誤差 (RFCM Phase 2 Exponential Error)
    # J 正比於 1 - exp(-omega^2 * d^2)
    # 設定 omega平方 = 0.5 讓曲線在圖表中好讀
    omega_sq = 0.5
    y_rfcm = 1 - np.exp(-omega_sq * d**2)
    
    # --- 開始繪圖 ---
    plt.figure(figsize=(10, 6))
    
    # 畫出兩條線
    plt.plot(d, y_euclidean, 'b--', linewidth=2, label=r'Traditional: $J \propto d^2$ (Unbounded / Rubber Band)')
    plt.plot(d, y_rfcm, 'g-', linewidth=4, label=r'RFCM Phase 2: $J \propto 1 - \exp(-\omega^2 d^2)$ (Bounded / Magnet)')
    
    # 畫出天花板線 (Saturation Level)
    plt.axhline(y=1.0, color='r', linestyle=':', alpha=0.6, label='Max Penalty Limit = 1.0')
    
    # --- 加入註解 (Annotations) 讓學生秒懂 ---
    
    # 註解 1: 傳統算法的無限懲罰
    plt.annotate('Penalty grows INFINITELY\n(Outliers pull heavily)', 
                 xy=(2.8, 2.8**2), xytext=(0.2, 5),
                 arrowprops=dict(facecolor='blue', shrink=0.05),
                 fontsize=12, color='darkblue', fontweight='bold',
                 bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="blue", alpha=0.9))
                 
    # 註解 2: RFCM 的停損點
    plt.annotate('Penalty STOPS growing\n(Outliers are ignored)', 
                 xy=(3.5, 1.0), xytext=(2.5, 4),
                 arrowprops=dict(facecolor='green', shrink=0.05),
                 fontsize=12, color='darkgreen', fontweight='bold',
                 bbox=dict(boxstyle="round,pad=0.3", fc="white", ec="green", alpha=0.9))

    # 設定圖表標籤
    plt.title('Cost Function Comparison', fontsize=16, fontweight='bold')
    plt.xlabel('Distance from Center ($d$)', fontsize=14)
    plt.ylabel('Error Cost / Penalty ($J$)', fontsize=14)
    plt.legend(fontsize=12, loc='best')
    plt.grid(True, alpha=0.3)
    
    # 限制顯示範圍，聚焦在差異處
    plt.ylim(0, 12)
    plt.xlim(0, 4.5)
    
    plt.tight_layout()
    
    # 儲存圖片
    filename = 'rfcm_phase2_function_comparison.png'
    plt.savefig(filename)
    print(f"圖表已生成並儲存為 {filename}")
    plt.show()

if __name__ == "__main__":
    plot_function_comparison()
```
