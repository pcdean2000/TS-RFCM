```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

def plot_hard_vs_fuzzy():
    # 
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))
    
    # 
    center_A = np.array([2, 5])
    center_B = np.array([8, 5])
    point_x = np.array([4.5, 5]) # 
    
    # 
    radius = 2.5
    
    # --- Hard Clustering (K-Means) ---
    ax1.set_title("Hard Clustering (K-Means)\nBinary Choice", fontsize=14, fontweight='bold')
    ax1.set_xlim(0, 10); ax1.set_ylim(0, 10)
    ax1.axis('off')
    
    # 
    circle1 = plt.Circle(center_A, radius, color='lightblue', alpha=0.3)
    circle2 = plt.Circle(center_B, radius, color='lightgreen', alpha=0.3)
    ax1.add_patch(circle1); ax1.add_patch(circle2)
    
    # 
    ax1.text(center_A[0], center_A[1], "Cluster A", ha='center', fontweight='bold')
    ax1.text(center_B[0], center_B[1], "Cluster B", ha='center', fontweight='bold')
    
    #  X
    ax1.scatter(point_x[0], point_x[1], s=150, c='blue', marker='o', edgecolors='black')
    ax1.text(point_x[0], point_x[1]-0.8, "Data x", ha='center')
    
    #  ( A A )
    ax1.plot([point_x[0], center_A[0]], [point_x[1], center_A[1]], 'b-', linewidth=3)
    ax1.text(3.2, 5.2, "u = 1", color='blue', fontweight='bold')
    ax1.text(6.5, 5.2, "u = 0", color='green', fontweight='bold')
    
    # --- Fuzzy Clustering (FCM) ---
    ax2.set_title("Fuzzy Clustering (FCM)\nShared Membership", fontsize=14, fontweight='bold')
    ax2.set_xlim(0, 10); ax2.set_ylim(0, 10)
    ax2.axis('off')
    
    # 
    circle3 = plt.Circle(center_A, radius, color='lightblue', alpha=0.3)
    circle4 = plt.Circle(center_B, radius, color='lightgreen', alpha=0.3)
    ax2.add_patch(circle3); ax2.add_patch(circle4)
    
    # 
    ax2.text(center_A[0], center_A[1], "Cluster A", ha='center', fontweight='bold')
    ax2.text(center_B[0], center_B[1], "Cluster B", ha='center', fontweight='bold')
    
    #  X ()
    ax2.scatter(point_x[0], point_x[1], s=150, c='purple', marker='o', edgecolors='black')
    ax2.text(point_x[0], point_x[1]-0.8, "Data x", ha='center')
    
    #  ()
    #  A ()
    ax2.plot([point_x[0], center_A[0]], [point_x[1], center_A[1]], 'b--', linewidth=2)
    ax2.text(3.2, 5.2, "u = 0.7", color='blue', fontweight='bold')
    
    #  B ()
    ax2.plot([point_x[0], center_B[0]], [point_x[1], center_B[1]], 'g--', linewidth=1)
    ax2.text(6.5, 5.2, "u = 0.3", color='green', fontweight='bold')
    
    # 
    ax2.text(5, 2, "Constraint: 0.7 + 0.3 = 1.0", ha='center', fontsize=12, 
             bbox=dict(boxstyle="round", fc="white", ec="black"))

    plt.tight_layout()
    plt.savefig('hard_vs_fuzzy_concept.png')
    plt.show()

if __name__ == "__main__":
    plot_hard_vs_fuzzy()
```
