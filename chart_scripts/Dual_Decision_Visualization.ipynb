```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch

def plot_dual_decision_boundary():
    # ... () ...
    np.random.seed(42)
    
    # 1. 
    # ---  (Normal) ---
    norm_a_d = np.random.normal(0.2, 0.1, 50)
    norm_a_s = np.random.normal(80, 5, 50) 
    norm_b_d = np.random.normal(0.3, 0.1, 30)
    norm_b_s = np.random.normal(60, 5, 30) 
    
    # ---  Type A:  ---
    anom_lone_d = np.random.normal(0.8, 0.1, 10)
    anom_lone_s = np.random.normal(5, 2, 10) 
    
    # ---  Type B:  ---
    anom_gang_d = np.random.normal(0.25, 0.05, 10)
    anom_gang_s = np.random.normal(10, 2, 10) 
    
    # 2. 
    plt.figure(figsize=(10, 8))
    
    plt.scatter(norm_a_d, norm_a_s, c='blue', alpha=0.6, s=100, label='Normal A (Browsing)')
    plt.scatter(norm_b_d, norm_b_s, c='cyan', alpha=0.6, s=100, label='Normal B (Backup)')
    plt.scatter(anom_lone_d, anom_lone_s, c='red', marker='^', s=100, label='Anomaly Type A (Lone Wolf)')
    plt.scatter(anom_gang_d, anom_gang_s, c='orange', marker='s', s=100, label='Anomaly Type B (Organized Gang)')
    
    threshold_d = 0.6
    threshold_size = 30
    
    plt.axvline(x=threshold_d, color='gray', linestyle='--', alpha=0.5)
    plt.axhline(y=threshold_size, color='gray', linestyle='--', alpha=0.5)
    
    plt.text(0.85, 90, "Zone 1: Outliers\n(Caught by Vector)", 
             ha='center', fontsize=12, fontweight='bold', color='darkred',
             bbox=dict(boxstyle="round", fc="#ffcccc", alpha=0.8))
             
    plt.text(0.25, 20, "Zone 2: Micro-clusters\n(Caught by Matrix/Size)", 
             ha='center', fontsize=12, fontweight='bold', color='darkorange',
             bbox=dict(boxstyle="round", fc="#ffe5cc", alpha=0.8))
             
    plt.text(0.25, 70, "Zone 3: Normal Behavior\n(Multi-modal)", 
             ha='center', fontsize=12, fontweight='bold', color='darkblue',
             bbox=dict(boxstyle="round", fc="#cce5ff", alpha=0.8))
    
    plt.annotate('', xy=(0.6, 5), xytext=(0.4, 5), arrowprops=dict(arrowstyle='->', lw=2))
    plt.text(0.5, 2, "D Threshold", ha='center', fontsize=10)
    
    plt.annotate('', xy=(0.05, 30), xytext=(0.05, 40), arrowprops=dict(arrowstyle='->', lw=2))
    plt.text(0.05, 35, "Size Threshold", ha='right', fontsize=10, rotation=90)

    plt.title("Visualizing the Dual Decision Path", fontsize=16, fontweight='bold')
    plt.xlabel("Outlier Score D(o) (Distance from Center)", fontsize=12)
    plt.ylabel("Cluster Size (Cohesion Strength)", fontsize=12)
    plt.legend(loc='upper right')
    plt.grid(True, alpha=0.3)
    
    plt.tight_layout()
    plt.savefig('dual_decision_plot.png')
    plt.show()

def plot_decision_flowchart():
    """
     (Flowchart)
    """
    fig, ax = plt.subplots(figsize=(8, 8))
    ax.axis('off') # 
    
    # 
    box_style = dict(boxstyle='round,pad=0.5', fc='white', ec='black', lw=2)
    decision_style = dict(boxstyle='round,pad=0.5', fc='#e6f2ff', ec='blue', lw=2)
    anom_a_style = dict(boxstyle='round,pad=0.5', fc='#ffcccc', ec='darkred', lw=2)
    anom_b_style = dict(boxstyle='round,pad=0.5', fc='#ffe5cc', ec='darkorange', lw=2)
    normal_style = dict(boxstyle='round,pad=0.5', fc='#ccffcc', ec='darkgreen', lw=2)
    
    arrow_props = dict(arrowstyle='->', lw=2, color='black')
    
    # 1. Start Node
    ax.text(0.5, 0.9, "Start: Node $x$", ha='center', va='center', size=14, bbox=box_style)
    ax.annotate('', xy=(0.5, 0.82), xytext=(0.5, 0.88), arrowprops=arrow_props)
    
    # 2. Check 1 (D value)
    ax.text(0.5, 0.75, "Check 1:\nIs Outlier Score $D(o)$ High?", ha='center', va='center', size=12, bbox=decision_style)
    
    # Path: Yes (To Anomaly A)
    ax.annotate('Yes', xy=(0.8, 0.75), xytext=(0.65, 0.75), arrowprops=arrow_props, fontsize=12, color='red')
    ax.text(0.8, 0.75, "Anomaly Type A\n(Lone Wolf)", ha='left', va='center', size=12, bbox=anom_a_style)
    
    # Path: No (Down)
    ax.annotate('No', xy=(0.5, 0.62), xytext=(0.5, 0.68), arrowprops=arrow_props, fontsize=12, color='blue')
    
    # 3. Check 2 (Cluster Size)
    ax.text(0.5, 0.55, "Check 2:\nIs Cluster Size Small?", ha='center', va='center', size=12, bbox=decision_style)
    
    # Path: Yes (To Anomaly B)
    ax.annotate('Yes', xy=(0.8, 0.55), xytext=(0.65, 0.55), arrowprops=arrow_props, fontsize=12, color='red')
    ax.text(0.8, 0.55, "Anomaly Type B\n(Organized Gang)", ha='left', va='center', size=12, bbox=anom_b_style)
    
    # Path: No (Down)
    ax.annotate('No', xy=(0.5, 0.42), xytext=(0.5, 0.48), arrowprops=arrow_props, fontsize=12, color='blue')
    
    # 4. Normal
    ax.text(0.5, 0.35, "Normal Behavior\n(Multi-modal Safe)", ha='center', va='center', size=14, bbox=normal_style)
    
    plt.title("EAC Dual Decision Logic Flow", size=16, weight='bold', y=0.98)
    plt.tight_layout()
    plt.savefig('decision_flowchart.png')
    plt.show()

if __name__ == "__main__":
    print("Generating Scatter Plot...")
    plot_dual_decision_boundary()
    print("Generating Flowchart...")
    plot_decision_flowchart()
```
